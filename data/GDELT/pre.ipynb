```python
import pandas as pd
```

```python
def load_index(input_path):
    index, rev_index = {}, {}
    with open(input_path) as f:
        for i, line in enumerate(f.readlines()):        
            rel, id = line.strip().split("\t")
            index[rel] = id
            rev_index[id] = rel
    return index, rev_index
    
```

```python
e2id,id2e = load_index("entity2id.txt")
r2id,id2r = load_index("relation2id.txt")
```

```python
def _read_triplets(filename):
    with open(filename, 'r+') as f:
        for line in f:
            processed_line = line.strip().split('\t')
            yield processed_line,line
```

```python
import hashlib
cols = ["Actor1Name","Actor2Name","EventCode","timid","Md5_list"]
data = [cols]
input_path = ["train.txt","valid.txt","test.txt"]
for input in input_path:
    for triplet,l in _read_triplets(input):
        tem_data = [id2e[triplet[0].strip()],id2e[triplet[2].strip()],id2r[triplet[1].strip()],triplet[3].strip(),hashlib.md5(l.encode()).hexdigest()]
        data.append(tem_data)
with open("GDELT.csv", "w") as f:
    for line in data:
        f.write("\t".join(line)+'\n')
```
